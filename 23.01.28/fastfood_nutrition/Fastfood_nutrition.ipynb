```python
max_sodium_index = df['Sodium_mg'].idxmax()
product_with_highest_sodium = df.loc[max_sodium_index, 'Product']

print("Product with the highest sodium:", product_with_highest_sodium)
```

```python
df['sodium_percent'] = (df['Sodium_mg'] * 100) / df['per_serve_size']

result = df.groupby(['product','Company'])['Sodium_mg'].mean().reset_index()
result = result.sort_values(by='sodium_percent')

print(result)
```
