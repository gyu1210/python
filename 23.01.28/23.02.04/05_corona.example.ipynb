```python
import pandas as pd
```

```python
#코로나 데이터 로드
corona = pd.read_csv("../../csv/corona.csv")
corona.head()
```

```python
# 컬럼을 삭제
# drop() 함수 이용 컬럼 삭제

### axis 라는 매개변수의 값을 기준으로 행(0), 열(1) 지정
# inplace 매개변수의 값을 기준으로 기준이 되는 데이터를 변경할지 지정 


corona.drop(["Unnamed: 0"], axis=1) #Unnamed: 0은 칼럼 이름
corona.drop(["Unnamed: 0"], axis=1, inplace=True) # 데이터를 변경하겠다 ; inplace=True
# =corona. corona.drop([Unnamed:0], axis=1)

```

```python
# 컬럼의 이름을 변경
# 컬럼의 목록 출력 방법
corona.columns = ["등록일시", "총사망자", "총확진자", "게시글번호", "기준일", "기준시간", "수정일시", "누적의심자", "누적확진률" ]
```

```python
corona.head(1)
```

```python
corona.drop(["기준일", "기준시간"], axis=1)
```

```python
#데이터프레임의 정렬을 변경
## 등록일시를 기준으로 오름차순
## sort_values([기준이 되는 컬럼명])

corona.sort_values(["등록일시"])

```

```python
#ascending 매개변수의 값이 true(기본값)면 오름차순, flase면 내림차순
corona.sort_values(["등록일시"], inplace=True)
```

```python
## 인덱스 값을 초기화 #기존자료를 함부로0 지우지 않음 그냥 index라고 정보를 더 줌 
## reset_index()
#corona.reset_index(.drop(['index],axis=1))
corona.reset_index(drop=True, inplace=True)
```

```python
#index 부분 삭제
corona.reset_index(drop=True, inplace=True)
```

```python
# 결측치 조회
corona.info()
```

```python
#결측치의 조회2 -합계
corona.isna().sum()
```

```python
corona.head()
```

```python
##데이터간 차이를 출력하는 함수
## diff(n) : n은 인덱스의 이동횟수 (n은 기본값이 1) ;1칸만 이동한다 ;차이
corona['총확진자'].diff()
```

```python
corona['일일확진자'] = corona['총확진자'].diff()
```

```python
corona.head()
```

```python
corona['총사망자'].diff()
```

```python
#일일사망자 컬럼 생성
corona['일일사망자'] = corona['총사망자'].diff()
```

```python
corona.head()
```

```python
## 인덱스를 이동
## shift(n) n만큼 인덱스 이동

corona['총사망자'].shift(1)
```

```python
corona['총사망자']
```

```python
# 결측치를 다른 값으로 대체
# fillna(n) : 결측치를 n으로 채워준다

##일일확진자, 일일사망자의 결측치를 0으로 대체

corona[['일일확진자', '일일사망자']].fillna(0) #결측치 0으로 대체
```

```python
#변경된 데이터 기존 데이터에 추가
corona[['일일확진자', '일일사망자']] = corona[['일일확진자', '일일사망자']] .fillna(0)

```

```python
corona.head()
```

```python
## fillna() method 매개변수 -> ffill (결측치 앞부분에 있는 데이터를 뒤에 채워줌- nan일때), bfill

corona['누적의심자'] = corona['누적의심자'].fillna(method='ffill')
```

```python
# 산점도 그래프
import numpy as np
```

```python
x = np.random.rand(50) #랜덤이기 때문에 매번 값이 다르게 나옴
y = np.random.rand(50)
```

```python
# 산점도 그래프 #plt.scatter
plt.scatter(x,y)
plt.show()
```

```python
## 영역, 색, 투명도 변경
color = np.random.rand(50)
area = (20* np.random.rand(50))**2 ##크기는 맘대로, 점크기가 작을까봐 제곱

plt.scatter(x,y,s = area, c = color, alpha = 0.5) #alpha 투명도 #s라는 매개변수와 c라는 매개변수는 공식임 s=color / c=area로 될수있음
plt.show()
```
